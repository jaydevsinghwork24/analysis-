```python
# copying some column from Courier Company - Invoice to my new DataFrame(df7)
df7=pd.DataFrame(df3[['Order ID','AWB Code','Billing Amount (Rs.)','Charged Weight','Customer Pincode','Zone','Type of Shipment']])
df7
```

```python
# renaming column and assigning it to df9
df9=df7.rename(columns={'Zone':'Zone_courrier',})
```

```python
# copying zone data from Company X - Pincode Zones to my DataFrame(df9)
df9['Zone_company']=df5['Zone']
df9
```

```python
# merging DataFrame of Company X - Order Report and Company X - SKU Master if column SKU matches and assign it to df10
df10 = pd.merge(df1, df4,how='right',on='SKU').reindex(df1.index)
df10
```

```python
# droping SKU from df10
df10.drop(['SKU'],inplace=True, axis=1)
df10
```

```python
#renaming column name
df10.rename(columns={'ExternOrderNo': 'Order ID',}, inplace=True)
df10
```

```python
# grouping our DataFrame by order id since company invoice carry only unique order id
dfa=df10.groupby(df10['Order ID']).sum()
dfa
```

```python
# creating  a new DataFrame from dfa and assigning my index value as one of column in it
df12=pd.DataFrame(dfa)
df12['Order ID']=df12.index
df12
```

```python
# merging weight column from df12 to df9 if order id matches
dfb=pd.merge(df9,df12[['Weight (g)']], how='left',on='Order ID')
dfb
```

```python
# converting weight in gram to kilogram
dfe=dfb
dfe['Weight (g)']=dfe['Weight (g)']/1000
```

```python
# rounding the decimal value in weight column
dfe['Weight (g)'] = dfe['Weight (g)'].round(2)
dfe
```

```python
# creating two empty column
dfe.insert(4,"wt slab by courrier", " ")
dfe.insert(10,"wt slab by company", " ")
```

```python
# saving dataframe as excel file
writer = pd.ExcelWriter('courrier.xlsx', engine='xlsxwriter')
dfe.to_excel(writer, sheet_name='Sheet1')
workbook  = writer.book
worksheet = writer.sheets['Sheet1']
# inserting data in empty columns with celling values of weight
worksheet.write_array_formula('F2:F125','=CEILING(E2:E125,0.5)') 
worksheet.write_array_formula('L2:L125','=CEILING(K2:K125,0.5)')
writer.close()
```

```python
for j in df6.index:
    for i in df8.index:
        if df8.loc[i, 'Zone_company']==df6.loc[j, 'zone']:
            if df8.loc[i, 'Type of Shipment']=='Forward charges':
                n=(df8.loc[i, 'wt slab by company'])/0.5
                if n<=1:
                    a=df6.loc[j, 'forward']
                    df8.loc[i, 'company_charge']=a
                else:
                    b=((n-1)*(df6.loc[j, 'additional']))+(df6.loc[j, 'forward'])
                    df8.loc[i, 'company_charge']=b
            elif df8.loc[i, 'Type of Shipment']=='Forward and RTO charges':
                n=(df8.loc[i, 'wt slab by company'])/0.5
                if n<=1:
                    a=(df6.loc[j, 'forward'])+(df6.loc[j, 'rto'])
                    df8.loc[i, 'company_charge']=a
                else:
                    b=((n-1)*(df6.loc[j, 'additional']))+(df6.loc[j, 'forward'])+((n-1)*(df6.loc[j, 'radditional']))+(df6.loc[j, 'rto'])
                    df8.loc[i, 'company_charge']=b
```
